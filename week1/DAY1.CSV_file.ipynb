```python
from pyspark.sql import session

from pyspark.sql.types import *
```

```python
df=spark.read.csv("/Volumes/workspace/default/day1pratice/C1.csv",header=True,inferSchema=True)
```

```python
df.show()
```

```python
df=df.replace(["blank","Blank"],None)
df.show()
```

```python
df=df.fillna({
    "Category":"Unknown",
    "Sales":"0",
    "JoinDate":"01-01-2023"
})
df.show()
```

```python
from pyspark.sql.functions import col,upper
df=df.withColumn("Country",upper(col("Country")))
df=df.replace("NEW YORK","USA")

df.show()
```

```python
df=df.dropDuplicates(["Name"])
df.show()
```
